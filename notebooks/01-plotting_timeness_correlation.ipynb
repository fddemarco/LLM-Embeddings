```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import polars as pl
import seaborn as sns

import constants
import experiments
import data_loader
import plots
```

```python
k_nns = (1, 3, 5)
```

```python
dfs = []
for k in k_nns:
    for model in constants.MODELS:
        for party in constants.PARTIES:
            embeddings = data_loader.load_sotu_data(party, model)
            df = experiments.mean_experiment(k, party,model, embeddings)
            dfs.append(df)
data = pd.concat(dfs)
data.to_parquet("timeness.parquet", index=False)
```

```python
def plot_correlation(dataset, x="Date", y="Timeness"):
    dataset.plot(x=x, y=y, marker="o", linestyle="-", color="b")
    correlation = dataset[y].corr(dataset[x], method="kendall")
    print(f"Correlation: {correlation:.2f}")
```
